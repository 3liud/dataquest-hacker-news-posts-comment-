## Hacker news posts exploration
The project aims to explore hacker news posts on Ask HN and Show HN to determine which of the two receives the most comments and other characterisits about them.

The project is based on Dataquest's practicval guided project on the python for data science course. 

```python
from csv import reader
dfile = ('../input/hacker-news-posts/HN_posts_year_to_Sep_26_2016.csv')
opened_file = open(dfile)
hn = reader(opened_file)
hn = list(hn)
```

Let's have a look at the data in the file. The data constitutes fo the following columns:

    id: The unique identifier from Hacker News for the post
    title: The title of the post
    url: The URL that the posts links to, if it the post has a URL
    num_points: The number of points the post acquired, calculated as the total number of upvotes minus the total number of downvotes
    num_comments: The number of comments that were made on the post
    author: The username of the person who submitted the post
    created_at: The date and time at which the post was submitted


```python
print(hn[:5])
```

exploring the header file and removing it from the dataset to make it possible for analysis

```python
header = hn[0]
hn = hn[1:]
print(hn[:5])
```

Separating the posts into differnt categories based on whther they are Ask HN, Show HN or general posts. 
then we count the number of posts in each category

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    if title.lower().startswith("ask hn"):
        ask_posts.append(row)
    elif title.lower().startswith("show hn"):
        show_posts.append(row)
    else:
        other_posts.append(row)
        
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
```

Calculating the Average Number of Comments for Ask HN and Show HN Posts

```python
total_ask_comments = 0

for row in ask_posts:
    counter = row[4]
    total_ask_comments += int(counter)
    
avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)
```

```python
total_show_comments = 0

for row in show_posts:
    counter = row[4]
    total_show_comments += int(counter)
    
avg_show_comments = total_show_comments / len(show_posts)
print(avg_show_comments)
```

From this analysis we can tell that the ask Hn posts receive the most comments than the show posts. this is based on the average seen on the analysis. the ask posts have an average comment of 10 while the show posts hve a lower average of 4 commnets per post. This implies that the ask hn posts receive the most comments and are somewhat more popular than the show posts. 

Finding the Amount of Ask Posts and Comments by Hour Created

```python
import datetime as dt

result_list = []

for post in ask_posts:
    created_time = post[6]
    n_comments = int(post[4])
    result_list.append([created_time, n_comments])
    
counts_by_hour = {} #number of ask posts
comments_by_hour = {} #corresponding number of comments in each ask post
date_format = "%m/%d/%Y %H:%M"

for row in result_list:
    date = row[0]
    comment = row[1]
    time = dt.datetime.strptime(date, date_format).strftime("%H")
    if time in counts_by_hour:
        comments_by_hour[time] += comment
        counts_by_hour[time] += 1
    else:
        comments_by_hour[time] = comment
        counts_by_hour[time] = 1

print(comments_by_hour)
print('\n')
print(counts_by_hour)
```

Calculating the Average Number of Comments for Ask HN Posts by Hour

```python
avg_by_hour = []

for hr in comments_by_hour:
    avg_by_hour.append([hr, comments_by_hour[hr] / counts_by_hour[hr]])
    
avg_by_hour
```

Sorting and Printing Values from a List of Lists

```python
swap_avg_by_hour = []

for row in avg_by_hour:
    swap_avg_by_hour.append([row[1], row[0]])
    
print(swap_avg_by_hour)

sorted_swap = sorted(swap_avg_by_hour, reverse=True)

sorted_swap
```

```python
print("Top 5 Hours for 'Ask HN' Comments")
for avg, hr in sorted_swap[:5]:
    print(
        "{}: {:.2f} average comments per post".format(
            dt.datetime.strptime(hr, "%H").strftime("%H:%M"),avg
        )
    )
```

The hour that receives the most comments per post on average is 15:00hr, with an average of 38.59 comments per post.
